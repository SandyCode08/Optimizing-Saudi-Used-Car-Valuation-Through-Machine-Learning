### **1. Business Problem Understanding**

#### **Context**

---

Arab Saudi sebagai salah satu negara dengan populasi besar di kawasan Timur Tengah memiliki dinamika pasar mobil bekas yang sangat aktif. Berdasarkan laporan dari Saudi Automobile Association dan data pasar otomotif regional, harga mobil bekas di Saudi terus mengalami fluktuasi dari tahun ke tahun. Perubahan ini dipengaruhi oleh berbagai faktor, mulai dari merek dan model mobil, tahun produksi, jarak tempuh (mileage), kondisi fisik kendaraan, hingga faktor eksternal seperti harga bahan bakar, inflasi, dan kebijakan impor kendaraan.

Mobil di Saudi juga memiliki permintaan yang tinggi karena beberapa alasan: mobilitas masyarakat yang sangat bergantung pada kendaraan pribadi, kondisi geografis yang luas, hingga status Saudi sebagai pusat aktivitas bisnis dan pekerja asing. Banyak ekspatriat, pekerja migran, dan penduduk lokal yang mencari mobil bekas sebagai alternatif lebih terjangkau dibandingkan membeli mobil baru, sehingga pasar mobil bekas selalu kompetitif.

Dalam konteks jual-beli mobil bekas, salah satu tantangan terbesar bagi pemilik mobil (seller) maupun dealer adalah menentukan harga jual yang tepat. Jika harga dipasang terlalu tinggi dibandingkan dengan mobil serupa di pasar, kendaraan tersebut bisa sulit terjual dan bertahan lama di listing. Sebaliknya, jika harga terlalu rendah, penjual berisiko kehilangan potensi keuntungan yang signifikan.

Ada banyak faktor yang memengaruhi harga jual mobil bekas di Saudi, mulai dari spesifikasi teknis (tipe transmisi, kapasitas mesin, konsumsi bahan bakar), kondisi fisik (interior dan eksterior), faktor merek (brand value dan popularitas di Saudi), hingga tren pasar otomotif global. Oleh karena itu, pemahaman yang komprehensif mengenai faktor-faktor ini sangat penting bagi penjual maupun dealer untuk menentukan harga yang optimal dan meningkatkan peluang mobil cepat terjual dengan keuntungan maksimal.

#### **Problem Statement**

---

Salah satu tantangan utama dalam pasar mobil bekas di Arab Saudi adalah menentukan harga jual mobil yang sesuai dengan kondisi pasar. Penetapan harga yang tepat menjadi hal yang sangat krusial, baik bagi pemilik mobil maupun dealer, karena harga yang terlalu tinggi akan menyulitkan mobil untuk terjual, sementara harga yang terlalu rendah akan mengakibatkan potensi kerugian finansial.

Meskipun data mengenai karakteristik mobil (seperti merek, model, tahun produksi, tipe transmisi, kapasitas mesin, jarak tempuh, dan kondisi kendaraan) tersedia, kompleksitas pasar otomotif membuat penentuan harga tidak bisa hanya mengandalkan intuisi atau perbandingan sederhana dengan mobil serupa. Dengan semakin bertumbuhnya jumlah mobil bekas yang dijual di pasar Saudi, diperlukan sebuah pendekatan berbasis data untuk dapat memprediksi harga mobil secara lebih akurat berdasarkan variabel-variabel atau fitur yang tersedia dalam dataset.

#### **Goals**

---

Berdasarkan permasalahan tersebut, penjual mobil maupun dealer di Arab Saudi membutuhkan sebuah tool yang dapat membantu mereka memprediksi harga jual mobil bekas secara akurat berdasarkan karakteristik kendaraan yang dimiliki. Dengan mempertimbangkan berbagai fitur yang tersedia dalam dataset, seperti merek, model, tahun produksi, tipe transmisi, kapasitas mesin, jarak tempuh, serta kondisi mobil, prediksi harga dapat lebih realistis dan sesuai dengan tren pasar otomotif di Saudi.

Bagi pemilik mobil, adanya prediksi harga yang akurat tentu dapat membantu mereka dalam menetapkan harga yang kompetitif, sehingga mobil dapat lebih cepat terjual tanpa harus kehilangan potensi keuntungan. Bagi dealer, alat prediksi ini dapat meningkatkan efisiensi kerja, memperkuat kepercayaan pelanggan, serta memberikan nilai tambah dalam proses jual-beli mobil bekas.

Dengan kata lain, semakin akurat prediksi harga mobil bekas, semakin besar peluang terjadinya transaksi yang menguntungkan bagi penjual, pembeli, maupun dealer, serta dapat berkontribusi pada transparansi dan stabilitas pasar otomotif di Arab Saudi.

#### **Analytic Approach**

---

Langkah yang perlu dilakukan adalah menganalisis data mobil bekas di Arab Saudi untuk menemukan pola dan hubungan antara berbagai fitur kendaraan, seperti merek, model, tahun produksi, tipe transmisi, kapasitas mesin, jarak tempuh (mileage), serta kondisi fisik mobil. Analisis ini bertujuan untuk memahami faktor-faktor apa saja yang paling berpengaruh terhadap harga mobil bekas.

Selanjutnya, akan dibangun sebuah model regresi yang dapat memprediksi harga mobil berdasarkan fitur-fitur tersebut. Model ini diharapkan dapat menjadi sebuah tool yang membantu pemilik mobil maupun dealer dalam menentukan harga jual mobil bekas yang lebih akurat, kompetitif, dan sesuai dengan kondisi pasar otomotif di Arab Saudi.

#### **Metric Evaluation**

---

Evaluasi model prediksi harga mobil bekas akan menggunakan tiga metrik utama, yaitu MSE (Mean Squared Error), RMSE (Root Mean Squared Error), dan R-squared (R-squared).

1. **MSE (Mean Squared Error)**
Mengukur rata-rata selisih kuadrat antara nilai aktual dengan nilai prediksi. MSE memberikan penalti lebih besar terhadap error yang besar, sehingga berguna untuk menilai seberapa jauh prediksi model dari nilai sebenarnya.

2. **RMSE (Root Mean Squared Error)**
Merupakan akar dari MSE yang mengembalikan satuan ke bentuk harga (misalnya Riyal). RMSE memudahkan interpretasi karena menggambarkan rata-rata besarnya kesalahan prediksi dalam satuan yang sama dengan target.

3. **R-squared (R²)**
Menunjukkan seberapa besar variasi harga mobil yang dapat dijelaskan oleh model berdasarkan fitur-fitur seperti merek, tahun, jarak tempuh, dan lainnya. Nilai R² yang mendekati 1 menandakan model memiliki kemampuan prediksi yang baik.

Dengan demikian, kombinasi **MSE, RMSE, dan R-squared** akan digunakan untuk mengevaluasi performa model regresi dalam memprediksi harga mobil bekas di Arab Saudi.

#### **Pipeline Imputer,Encoder,Scaler**

---

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import FunctionTransformer

#One Hot Encoder Kita buat dalam pipeline
one_hot_transformer = Pipeline(steps= [
    ('imputer', SimpleImputer(strategy='constant', fill_value='Missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])


#Standard Scaler Kita buat dalam pipeline
standard_scaler_transformer = Pipeline(steps=[
     ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

label_transformer = Pipeline(steps = [
    ('to_int', FunctionTransformer(lambda x: x.astype(int))),  # True/False -> 1/0
    ('imputer', SimpleImputer(strategy='most_frequent'))
])

# kita buat pipelinenya untuk column transformer
preprocessor =  ColumnTransformer(
    transformers = [
        ('scaler',standard_scaler_transformer,numerical_cols),
        ('onehot',one_hot_transformer,one_hot_cols),
        ('label', label_transformer, boolean_cols)

    ]
)

preprocessor
```

##### **Weighted Averaging**

Pada tahap ini, dilakukan penggabungan dua model terbaik, yaitu Random Forest dan XGBoost, menggunakan metode Weighted Averaging.

Metode ini bekerja dengan memberikan bobot (weight) pada hasil prediksi setiap model berdasarkan performanya.
Model dengan performa lebih baik (R² lebih tinggi) akan mendapat bobot lebih besar dalam perhitungan prediksi akhir.

##### **Stacking (Meta-Model)**

---

Setelah memperoleh dua model dengan performa terbaik sebelumnya — yaitu XGBoost dan Random Forest, dilakukan pendekatan ensemble learning dengan teknik stacking untuk menggabungkan kekuatan keduanya.

**Konsep Stacking**

Stacking merupakan metode ensemble di mana beberapa model dasar (base learners) digabungkan dengan model tingkat akhir (meta-model).

**Pada tahap ini:**

**1.Base model:**

- XGBoost (model_pipeline_tuningnew_xgboost)

- Random Forest (model_pipeline_new_randomforest)

**2. Meta model (final estimator):**

- Ridge Regression dengan cross-validation otomatis menggunakan RidgeCV(alphas=[0.1, 1, 10], cv=5)

- Model meta dilatih menggunakan prediksi dari dua model dasar sebagai fitur input, sehingga mampu mempelajari kapan harus “mempercayai” XGBoost atau Random Forest untuk menghasilkan prediksi akhir yang lebih stabil.

###### **Stacking Model Calculation**

```python
from sklearn.ensemble import StackingRegressor
from sklearn.linear_model import RidgeCV

estimators = [
    ('xgb', model_pipeline_tuningnew_xgboost),
    ('rf', model_pipeline_new_randomforest)
]

stacking_model = StackingRegressor(
    estimators=estimators,
    final_estimator=RidgeCV(alphas=[0.1, 1, 10], cv=5)
)

# Fit di train
stacking_model.fit(X_train_new, y_train_new)

# Prediksi
y_pred_stack = stacking_model.predict(X_test_new)

# Evaluasi
r2_stack = r2_score(y_test_new, y_pred_stack)
mse_stack = mean_squared_error(y_test_new, y_pred_stack)
rmse_stack = mse_stack ** 0.5


best_results['Stacking (Meta Model)'] = {
    'R-squared': r2_stack,
    'RMSE': rmse_stack 
}

print(f"R2 Stacking: {r2_stack:.4f}")
print(f"MSE Stacking: {mse_stack:.2f}")
print(f"RMSE Stacking: {rmse_stack:.2f}")
```

###### **Insight Stacking Model**

Pendekatan stacking ensemble berhasil diterapkan dengan menggabungkan dua model terbaik (XGBoost dan Random Forest) menggunakan meta-model RidgeCV.

Namun, hasil R-squared model stacking sedikit lebih rendah dibanding XGBoost, yang menandakan bahwa XGBoost sudah mampu menangkap pola data dengan baik tanpa perlu kombinasi tambahan.

##### **Model**

```python
from sklearn.preprocessing import FunctionTransformer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from xgboost import XGBRegressor
```

```python
## kita pastiin top 15 make dan 20 type
TOP_15_MAKE = data_saudi_copy['Make'].value_counts().nlargest(15).index.tolist()
TOP_20_TYPE = data_saudi_copy['Type'].value_counts().nlargest(20).index.tolist()
```

```python
# Kita Buat fungsi Yang otomatis Membuatt Kolom Type and Make Reduced, seta membuat kolom bolean jadi integer

def reduce_make(data_saudi_feature_selection):
    data_saudi_feature_selection['Make_reduced'] = data_saudi_feature_selection['Make'].where(data_saudi_feature_selection['Make'].isin(TOP_15_MAKE), 'Other')
    return data_saudi_feature_selection[['Make_reduced']]

def reduce_type(data_saudi_feature_selection):
    data_saudi_feature_selection['Type_reduced'] = data_saudi_feature_selection['Type'].where(data_saudi_feature_selection['Type'].isin(TOP_20_TYPE), 'Other')
    return data_saudi_feature_selection[['Type_reduced']]

def to_int_transform(x):
    return x.astype(int)
```

```python
### Kita Buat Preprocessor yang final biar enak bisa langsung dipakai


one_hot_transformer = Pipeline(steps= [
    ('imputer', SimpleImputer(strategy='constant', fill_value='Missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])


#Standard Scaler Kita buat dalam pipeline
standard_scaler_transformer = Pipeline(steps=[
     ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

label_transformer = Pipeline(steps = [
    ('to_int', FunctionTransformer(to_int_transform)),  # True/False -> 1/0
    ('imputer', SimpleImputer(strategy='most_frequent'))
])

make_transformer = Pipeline([
    ('reduce', FunctionTransformer(reduce_make, validate=False)),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

type_transformer = Pipeline([
    ('reduce', FunctionTransformer(reduce_type, validate=False)),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
# Hapus kolom Make dan Type dari one_hot_cols
one_hot_cols_final = [col for col in one_hot_cols if col not in ['Make', 'Type']]


preprocessor_final = ColumnTransformer([
    ('scaler', Pipeline([('imputer', SimpleImputer(strategy='median')), ('scaler', StandardScaler())]), numerical_cols),
    ('onehot', Pipeline([('imputer', SimpleImputer(strategy='constant', fill_value='Missing')), ('onehot', OneHotEncoder(handle_unknown='ignore'))]), one_hot_cols_final),
    ('label', label_transformer, boolean_cols),
    ('make', make_transformer, ['Make']),
    ('type', type_transformer, ['Type'])
    ])


#preprocessor_final = ColumnTransformer(
#    transformers=[
#        ('scaler', standard_scaler_transformer, numerical_cols),
#        ('onehot', one_hot_transformer, one_hot_cols_final),
#        ('label', label_transformer, boolean_cols),
#        ('make', make_transformer, ['Make']),
#        ('type', type_transformer, ['Type'])
#    ]
#)


```

```python
model_pipeline_xgboost_final = Pipeline(steps=[
    ('preprocessor', preprocessor_final),
    ('xgboost', XGBRegressor(
        random_state=42,
        learning_rate=0.1,
        max_depth=5,
        n_estimators=100
        ))
    
])

model_pipeline_xgboost_final
```
